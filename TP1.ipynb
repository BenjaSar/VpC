### Coordenadas cromaticas

```python
# Cargamos la imagen 1
img1 = cv.imread('././imgs/coord_cromaticas/CoordCrom_1.png')
imgRGB_1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)

plt.imshow(img1)
plt.show()

# Dimensi칩n de la imagen 1
print(f'Tama침o de la imagen uno: {img1.shape} pix')
```

```python
#Function for Chromaticity coordinates

h = img1.shape[0]
w = img1.shape[1]

def chromaticity_coordinates(input_image:np.ndarray) -> np.ndarray:
    img_output = np.zeros(img1.shape, np.float16)
    input_image = input_image.astype(np.float16)
    
    #Loop through every pixel column
    for i in range(h):
        #Loop through every pixel row
        for j in range(w):
          if (input_image[i,j] == 0).all():
               img_output[i,j] = 0
          else:
               r = input_image[i,j][0]
               g = input_image[i,j][1]
               b = input_image[i,j][2]
               r_output = r/(r+g+b)
               g_output = g/(r+g+b)
               b_output = b/(r+g+b)
               img_output[i,j] = (r*r_output, g*g_output, b*b_output)
        
        img_output = img_output.astype(np.uint8)
    plt.imshow(img_output)
    plt.show()
    return img_output  
    
     
```
